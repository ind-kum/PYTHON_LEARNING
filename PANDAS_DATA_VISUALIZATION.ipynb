These methods can be provided as the kind keyword argument to plot(). These include-

bar or barh for bar plots

hist for histogram

box for boxplot

'area' for area plots

'scatter' for scatter plots

```python
# BAR PLOT
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])

print(df)

df.plot.bar()
```

```python
import pandas as pd
df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
df.plot.bar(stacked=True)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])

df.plot.barh(stacked=True)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'a':np.random.randn(1000)+1,'b':np.random.randn(1000),'c':
np.random.randn(1000) - 1}, columns=['a','b','c'])
 
df.plot.hist(bins=20)
```

Box Plots Boxplot can be drawn calling Series.box.plot() and DataFrame.box.plot(),or DataFrame.boxplot() to visualization the distribution of values within each column,

For instance, here is a boxplot representing five trials of 10 observations of a uniform random variable on[0,1]

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.rand(10,5), columns=['A','B','C','D','E'])
df
df.plot.box()
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(10, 4), columns=['a','b','c','d'])
df.plot.area()
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.rand(50,4), columns=['a','b','c','d'])
df.plot.scatter(x='a', y='b')
```
