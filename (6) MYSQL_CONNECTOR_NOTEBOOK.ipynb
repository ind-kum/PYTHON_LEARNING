```python
import pandas as pd
con=mysql.connector.connect(
    host="localhost",
    port=3306,
    user="root",
    password="1234",
    database="world")
```

```python
df=pd.read_sql_query("Select * from country",con)
df
df.head(10)
```

```python
def Q(sql):
    res=pd.read_sql_query(sql,con)
    return (res)
```

```python
res=Q("select * from country")
res
```

```python
res=Q("select * from country where name like 'A%'")
res
```

```python
res.columns
```

```python
res=Q('select count(code) from country')
res
```

```python
res=Q('select distinct(continent)from country')
res
```

```python
res=Q('select capital,continent from country group by continent')
res
```

```python
res=Q('select count(population),continent from country group by continent')
res
```

```python
res=Q('select sum(population) from country')
res
```

```python
res=Q('select avg(population) from country')
res
```

```python
res=Q('select count(governmentform) from country')
res
```

```python
res=Q('select count(governmentform),continent from country group by continent')
res
```

```python
res=Q('select count(governmentform),continent,code2 from country group by continent order by code2')
res
```
