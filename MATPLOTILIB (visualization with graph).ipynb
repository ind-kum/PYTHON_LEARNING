Histogram


A histogram takes in a series of data and  divides the data into a number of bins. it then plots the frequency data points in each bin (i.e. the interval of point). It is usefull in understanding the count of data ranges. When to use: We should use histogram when we need the count of the variable in plot.

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a, bins = [0,20,40,60,80,100])
plt.title("histogram")
plt.show()
```
